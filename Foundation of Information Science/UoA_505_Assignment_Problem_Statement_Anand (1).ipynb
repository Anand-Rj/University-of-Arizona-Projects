
> 2.b  Provide the following statistics for query string “avatar movie” of top 50 videos sorted by relevance in the US region ( 7 marks )

> Output expected: video ID, title, no of views, no of likes,no of comments exported to CSV file






```python
#Function to get collect the data of the Avatar movie data

def id_collection(youtube):

    request = youtube.search().list(
                part="snippet",
                type= "video",
                maxResults=50,
                q="avatar movie"
                )

    video_ids = []

    response = request.execute()

    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['id']['videoId'])

    return video_ids
```

```python
#list of video id
id_collection(youtube)

video_id = id_collection(youtube)
video_id
```

Reference link: https://developers.google.com/youtube/v3/docs/videos/list

```python
#Function to get the stats of the top 50 videos in the US region

def get_stats(youtube, video_id):

        stats = []

        request = youtube.videos().list(
                        part="snippet,statistics",
                        id =video_id,
                        )

        response = request.execute()

        for video in response['items']:
            stats_desc = dict(video_ID = video['id'],
                              Title = video['snippet']['title'],
                              No_of_Views = video['statistics'].get('viewCount'),
                              No_of_likes = video['statistics']['likeCount'],
                              No_of_Comments = video['statistics']['commentCount']
                              )
            stats.append(stats_desc)

        return stats
```

```python
#filtered out video statistics
video_stats = get_stats(youtube, video_id)
video_stats
```

```python
#converting the statistics and saving it as csv file
video_data = pd.DataFrame(video_stats)

video_data.to_csv('./videostatistics.csv', index=False)
print("Saved Successfully!")
```
